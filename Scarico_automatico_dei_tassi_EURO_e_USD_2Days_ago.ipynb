## ***SCARICO CURVE EURO***

```python
df = pd.read_csv('/content/EUREuropeIndustrial.csv', parse_dates=['PX_DT_2D'])
```

```python
df['LAST_UPDATE_DT'] = df['PX_DT_2D'].dt.strftime('%d-%m-%y')
```

```python
df['LAST_UPDATE_DT'] = df['LAST_UPDATE_DT'].astype(str) + ' EUR'
```

```python
df["PX_LAST_EOD"] = df["PX_CLOSE_2D"].astype(str)
```

```python
df['PX_LAST_EOD'] = df['PX_LAST_EOD'].apply(lambda x: x.replace('.', ','))
```

```python
df
```

**Modalità di scarico delle curve con divisa EURO -- RATING A-**



```python
DF = df[df['IDENTIFIER'].str.startswith('BVCWFJ')]
```

```python
DF
```

```python
Data = DF.drop(columns=["DL_REQUEST_ID", "DL_REQUEST_NAME", "DL_SNAPSHOT_TZ", "DL_SNAPSHOT_START_TIME", "RC", "IDENTIFIER", "DL_SNAPSHOT_START_TIME","LAST_UPDATE_DT", "PX_CLOSE_2D", "PX_DT_2D"])
```

```python
DATA = Data.transpose()
```

```python
DATA
```

```python
file_name = 'tassi-Ameno-EUR.xlsx'
```

```python
DATA.to_excel(file_name)
print('DataFrame is written to Excel File successfully.')
```

```python
# opening the source excel file
wbo = openpyxl.load_workbook(file_name)
```

```python
#attach the ranges to the sheet
wso = wbo["Sheet1"]["B2":"P2"]
```

```python
# opening the destination excel file
wbd = openpyxl.load_workbook("/content/Condizioni economiche Fin. IC BREVE E MLT Eni II trimestre 2024.xlsx", keep_vba=True)
```

```python
tab_name = df.at[0, 'LAST_UPDATE_DT']
```

```python
#attach the ranges to the sheet
wsd = wbd[tab_name]["C4":"Q4"]
```

```python
#step1 : pair the rows
for row1,row2 in zip(wso,wsd):
    #within the row pair, pair the cells
    for cell1, cell2 in zip(row1,row2): cell2.value = cell1.value
```

```python
#save document
wbd.save('Condizioni economiche Fin. IC BREVE E MLT Eni II trimestre 2024.xlsm')
```
